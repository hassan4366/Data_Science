Filtering Rows Based on Conditions:

```python
df['bedrooms']
```

```python
# Filtering rows where 'bedrooms' is greater than or equal to 3
filtered_rows = df[df['bedrooms'] >= 3]
(filtered_rows.head())
```

```python
# Combining multiple conditions using logical operators (& for AND, | for OR)
filtered_rows1 = df[(df['bedrooms'] >= 3) & (df['bathrooms'] >= 2)]
(filtered_rows1.head())
```

```python
df[(df['bedrooms'] >= 3) & (df['bathrooms'] >= 2)].sort_values(by='bathrooms') #         
```

Grouping and Aggregating Data:
Pandas groupby is used for grouping the data according to the categories and apply a function to the categories. It also helps to aggregate data efficiently.

```python
# Grouping by 'bedrooms' and calculating mean price and max sqft_living
grouped_data = df.groupby('bedrooms').agg({'price': 'mean', 'sqft_living': 'max'}) 
grouped_data
```

crosstab()
This method is used to compute a simple cross-tabulation of two (or more) factors. By default, computes a frequency table of the factors unless an array of values and an aggregation function are passed

```python
pd.crosstab( df.price, df.bedrooms)
```

```python
pd.crosstab( df.price, df.bedrooms, normalize = "index" )
```

```python
pd.crosstab( df.price, df.bedrooms, normalize = "columns" )
```

```python
pd.crosstab( df.price, df.bedrooms, normalize = "all" )
```
