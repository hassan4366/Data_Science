```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart_2020_cleaned.csv')
df.head(10)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x="GenHealth",data=df)
plt.title("General Health category")
plt.show
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='BMI',y='GenHealth',hue='Sex',data=df)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='SleepTime', y='GenHealth',data=df)
```

```python
sns.boxplot(x='SleepTime',y='Diabetic',data=df,showmeans=True)
```

```python
sns.boxplot(x='BMI',hue='AgeCategory',data=df)
```

```python
sns.violinplot(x='BMI',hue='Sex',data=df)
```

```python
from numbers import Rational
sns.violinplot(x='Diabetic',y='SleepTime',data=df)
```

```python
sns.stripplot(x="AgeCategory", y="SleepTime", data=df, jitter=True)
plt.xticks(rotation=45)
```

```python
sns.swarmplot(x="Sex", y="BMI", hue="Diabetic", data=df)
```

```python
sns.catplot(x="Sex", y="SleepTime", col="GenHealth", data=df, kind="box")
```

```python
sns.histplot( x="SleepTime",data=df, bins=20, kde=True)
```

```python
sns.displot(x="BMI",data=df,kde=True, bins=30)
```
